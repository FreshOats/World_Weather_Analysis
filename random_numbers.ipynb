```python
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x+=1
    return latitudes

%timeit latitudes(1500)
```

```python
def latitudes(size):
    latitudes = []
    for x in range(size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
    return latitudes


latitudes(1500)
```
