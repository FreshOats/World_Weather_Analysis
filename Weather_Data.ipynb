```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy
import time
from datetime import datetime
import requests
from config import weather_api_key
from scipy.stats import linregress

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
# 3. Create a set of 2000 random coordinates

lats = np.random.uniform(low=-90.000, high=90.000, size=100)
lngs = np.random.uniform(low=-180.000, high=180.000, size=100)
coordinates = list(zip(lats, lngs))




# 4. Find the nearest city to each of the sets of coordinates using citipy

cities = []

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name


    # Add only unique cities to the cities list
    if city not in cities:
        cities.append(city)

len(cities)
```

```python
# 5. Perform the API call with OpenWeatherMap

city_data = []

print("-----------------------------------------")
print("        Commencing Data Acquision        ")
print("-----------------------------------------")

# Set the counters - the API limits the number of calls over a period of time, so this will be run in sets of 50
record_count = 1
set_count = 1

# Loop through the cities in the list using enumerate to avoid nested loops for the cities

for i, city in enumerate(cities):

    # Group cities in sets of 50
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        time.sleep(1)

    # Create an endpoint URL with each activity
    city_url = url + "&q=" + city.replace(" ", "+")

    # Log the URL, record, set numbers, and the city
    print(f"Processing record {record_count} of Set {set_count} | {city}")

    # Add 1 to the record count
    record_count += 1

    # Run an API request for each of the cities.append
    try: 
        city_weather = requests.get(city_url).json()
        # Parse the Lat/Lng, Max temp, % humidity, % cloudiness, Wind speed, weather description
        city_country = city_weather["sys"]["country"]
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_description = city_weather["weather"]["description"]
        # Append the above info into the city_data list
        city_data.append({"City": city.title(),
                          "Country": city_country,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Current Description": city_description})
    except:
        print("City not found. Skipping...")
        pass

# Indicate that loading is complete
print("-----------------------------------------")
print("         Data Acquision Complete         ")
print("-----------------------------------------")
```

```python
# 7. Add the data to a new DataFrame
city_data_df = pd.DataFrame(city_data)
city_data
```

```python
# 8. Export the data to the Database folder
output_file = "Weather_Database/WeatherPy_Database.csv"

city_data_df.to_csv(output_file, index_label = "City_ID")
```
